```python
import pandas as pd
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('IBM-HR-Employee-Attrition.csv')
df.head()
```

```python
# Randomly select 100 rows from the DataFrame
random_rows = df.sample(n=100, random_state=42) 
```

```python
random_rows = random_rows.drop(columns=['Attrition'])

# Create a new DataFrame with the selected rows
new_df = pd.DataFrame(random_rows)

# Display the new DataFrame
new_df.head()
```

```python
attrition_counts = new_df['Attrition'].value_counts()

# Display the counts
print(attrition_counts)
```

```python
diff_count = attrition_counts['No'] - attrition_counts['Yes']

# Sample 'Yes' values to balance the counts
additional_yes_values = df[df['Attrition'] == 'Yes'].sample(n=diff_count, replace=True)

# Append sampled 'Yes' values to the DataFrame
df = pd.concat([df, additional_yes_values], ignore_index=True)

# Count the new distinct values under the 'Attrition' column
new_attrition_counts = df['Attrition'].value_counts()

# Display the new counts
print(new_attrition_counts)
```

```python
# Save the new DataFrame as "machine_Test_data.csv"
new_df.to_csv('machine_test_data.csv', index=False)
```
