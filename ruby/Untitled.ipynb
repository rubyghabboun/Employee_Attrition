```python
import pandas as pd

# Read the dataset
data = pd.read_csv('Resources/IBM-HR-Employee-Attrition.csv')

# Count the occurrences of each class in the Attrition column
attrition_counts = data['Attrition'].value_counts()

# Calculate the desired count for each class
desired_count_no = int(1471 * 0.6)  # 60% of total rows
desired_count_yes = int(1471 * 0.4)  # 40% of total rows

# Filter rows for each class and balance the data
data_no = data[data['Attrition'] == 'No'].sample(n=min(desired_count_no, attrition_counts['No']), random_state=42)
data_yes = data[data['Attrition'] == 'Yes'].sample(n=min(desired_count_yes, attrition_counts['Yes']), random_state=42)

# Concatenate the balanced data
balanced_data = pd.concat([data_no, data_yes])

# Shuffle the dataset
balanced_data = balanced_data.sample(frac=1, random_state=42)

# Optional: You can save the balanced dataset to a new CSV file if needed
# balanced_data.to_csv('balanced_dataset.csv', index=False)

# Now you have the balanced dataset with 60% 'No' and 40% 'Yes' Attrition
```
