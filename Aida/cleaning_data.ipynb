```python
# Import findspark and initialize. 
import findspark
findspark.init()
```

```python
# Import packages
import pyspark
from pyspark.sql import SparkSession
import pandas as pd
```

```python
# Create a SparkSession
spark = SparkSession.builder.appName("Attrition").getOrCreate()
```

```python
# Read in the AWS S3 bucket into a DataFrame.
from pyspark import SparkFiles
url = "https://thesharpies-032024.s3.amazonaws.com/IBM-HR-Employee-Attrition.csv"
spark.sparkContext.addFile(url)
df_spark = spark.read.csv(SparkFiles.get("IBM-HR-Employee-Attrition.csv"),sep=",", header=True, inferSchema=True)

# Show PySpark DataFrame
df_spark.show()
```

```python
# Print schema
df_spark.printSchema()
```

```python
# Show the columns
df_spark.columns
```

```python
# Converting PySpark DataFrame to Pandas DataFrame
df = df_spark.toPandas()

# Showing all columns of the dataframe
pd.set_option('display.max_columns', None)

# Display the first few rows of the DataFrame to understand its structure
df.head()
```

```python
# Check for missing values in the DataFrame
df.isnull().sum()
```

```python
# Data description for numerical columns
df.describe()
```

```python
# Data description for categorical columns
df.describe(include='object')
```

```python
# Unique values of categorical columns
df.select_dtypes('object').apply(pd.Series.unique)
```

```python
# Drop columns with no relevant data
columns_to_drop = ['DailyRate', 'EducationField', 'HourlyRate', 'JobLevel',
                   'MaritalStatus', 'StandardHours','StockOptionLevel']

# Drop the specified columns from the DataFrame
df = df.drop(columns=columns_to_drop)
```

```python
df.head()
```

```python
# Export the DataFrame to a CSV file
output_file_path = "IBM_clean_employee_data.csv"

df.to_csv(output_file_path, index=False)

print("CSV file exported successfully.")
```
