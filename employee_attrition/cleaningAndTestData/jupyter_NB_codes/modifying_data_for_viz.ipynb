```python
import pandas as pd

```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv("IBM-HR-Employee-Attrition.csv")

# Display the first few rows of the DataFrame to understand its structure
df.head()
```

```python
# Check for missing values in the DataFrame
print(df.isnull().sum())

# Check for any inconsistencies or strange values in each column
print(df.describe(include='all'))
```

```python
education_reverse_mapping = {1: 'Below College', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor'}
environment_satisfaction_reverse_mapping = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
job_involvement_reverse_mapping = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
job_satisfaction_reverse_mapping = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
performance_rating_reverse_mapping = {1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding'}
relationship_satisfaction_reverse_mapping = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
work_life_balance_reverse_mapping = {1: 'Bad', 2: 'Good', 3: 'Better', 4: 'Best'}

# Apply reverse mappings to the respective columns
df['Education'] = df['Education'].replace(education_reverse_mapping)
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].replace(environment_satisfaction_reverse_mapping)
df['JobInvolvement'] = df['JobInvolvement'].replace(job_involvement_reverse_mapping)
df['JobSatisfaction'] = df['JobSatisfaction'].replace(job_satisfaction_reverse_mapping)
df['PerformanceRating'] = df['PerformanceRating'].replace(performance_rating_reverse_mapping)
df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].replace(relationship_satisfaction_reverse_mapping)
df['WorkLifeBalance'] = df['WorkLifeBalance'].replace(work_life_balance_reverse_mapping)

# Display the DataFrame to verify the transformation
df.head()
```

```python
columns_to_drop = ['DailyRate', 'EducationField', 'HourlyRate', 'JobLevel',
                   'MaritalStatus', 'Over18', 'OverTime', 'StandardHours',
                   'StockOptionLevel', 'TrainingTimesLastYear']

# Drop the specified columns from the DataFrame
df = df.drop(columns=columns_to_drop)
```

```python
output_file_path = "modified_employee_attrition.csv"

# Export the DataFrame to a CSV file
df.to_csv(output_file_path, index=False)

print("CSV file exported successfully.")
```
